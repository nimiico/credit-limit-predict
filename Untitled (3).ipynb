```python
df = df.drop("Unnamed: 19", axis=1)
df = df.drop_duplicates(subset=["CLIENTNUM"], keep=False)
df = df.drop("CLIENTNUM", axis=1)
df = df.drop("CLIENTNUM", axis=1)
```

```python
df['Gender'] = LabelEncoder().fit_transform(df['Gender'])
df['Education_Level'] = LabelEncoder().fit_transform(df['Education_Level'])
df['Marital_Status'] = LabelEncoder().fit_transform(df['Marital_Status'])
df['Income_Category'] = LabelEncoder().fit_transform(df['Income_Category'])
df['Card_Category'] = LabelEncoder().fit_transform(df['Card_Category'])
```

```python
df = pd.DataFrame(df, columns=['Customer_Age', 'Gender', 'Dependent_count', 'Education_Level',
       'Marital_Status', 'Income_Category', 'Card_Category', 'Months_on_book',
       'Total_Relationship_Count', 'Months_Inactive_12_mon',
       'Contacts_Count_12_mon', 'Total_Revolving_Bal',
       'Total_Amt_Chng_Q4_Q1', 'Total_Trans_Amt', 'Total_Trans_Ct',
       'Total_Ct_Chng_Q4_Q1', 'Avg_Utilization_Ratio', 'Credit_Limit'])
```

```python
data = df.iloc[:, 0:-1]
label = df.iloc[:,-1]
```

```python
x_train, x_test, y_train, y_test = train_test_split(data, label, test_size=0.30, shuffle=True, random_state=7)
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
#scaler = StandardScaler()
scaler = scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

model = LinearRegression().fit(x_train, y_train)
score = model.score(x_test,y_test)
y_pred = model.predict(x_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
print(f'score: {score}')

```

```python
import numpy as np
import pandas as pd
from sklearn.impute import KNNImputer
from sklearn.preprocessing import StandardScaler, LabelEncoder

# Example dataset with both numeric and categorical features
data = {
    'Numeric1': [1, 2, np.nan, 4, 5],
    'Numeric2': [10, np.nan, 30, 40, 50],
    'Categorical': ['A', 'B', 'A', np.nan, 'B']
}

df = pd.DataFrame(data)

# Convert categorical feature to numerical using LabelEncoder
label_enc = LabelEncoder()
df['Categorical'] = label_enc.fit_transform(df['Categorical'].astype(str))

# Separate numeric and categorical features
numeric_cols = ['Numeric1', 'Numeric2']
categorical_cols = ['Categorical']

# Normalize numeric features
scaler = StandardScaler()
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

# Apply KNN imputation separately for numeric and categorical features
imputer_numeric = KNNImputer(n_neighbors=2)
imputed_numeric = imputer_numeric.fit_transform(df[numeric_cols])

imputer_categorical = KNNImputer(n_neighbors=2)
imputed_categorical = imputer_categorical.fit_transform(df[categorical_cols])

# Convert imputed data back to DataFrame
imputed_df = pd.DataFrame(np.concatenate((imputed_numeric, imputed_categorical), axis=1),
                          columns=numeric_cols + categorical_cols)

# Inverse transform for categorical feature
imputed_df['Categorical'] = label_enc.inverse_transform(np.round(imputed_df['Categorical']).astype(int))

# Inverse transform for numeric features


imputed_df[numeric_cols] = scaler.inverse_transform(imputed_df[numeric_cols])

print("Imputed DataFrame:")
print(imputed_df)
```

```python
import numpy as np
import pandas as pd
from sklearn.impute import KNNImputer
from sklearn.preprocessing import OneHotEncoder

# Create a sample dataset
data = {
    'Feature1': ['A', 'B', 'C', 'A', np.nan, 'B', 'C', 'A', 'B', 'C'],
    'Feature2': ['X', 'Y', np.nan, 'X', 'Y', 'Z', 'X', np.nan, 'Z', 'Y'],
    'Feature3': ['High', 'Low', 'Medium', 'Low', 'High', np.nan, 'Medium', 'High', 'Low', 'Medium']
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

# One-hot encode categorical features
df_encoded = pd.get_dummies(df)

# Apply the KNNImputer
knn_imputer = KNNImputer(n_neighbors=3, weights="uniform", metric='nan_euclidean')
df_imputed_encoded = knn_imputer.fit_transform(df_encoded)

# Convert back to DataFrame
df_imputed = pd.DataFrame(df_imputed_encoded, columns=df_encoded.columns)

print("\nDataFrame after imputation:")
df_imputed
```
